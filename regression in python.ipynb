```python
#Create variable with those values
m,c = np.polyfit(x, y, 1)
# we had a look at m and c
m, c
```

```python
# Plot x and y and the regression line in red.
plt.plot(x, y, 'k.')
plt.plot(x, m * x + c, 'b-')
```

We can calculate best of m and c

```python
# Calculate mean x and mean y.
x_avg = np.mean(x)
y_avg = np.mean(y)

# Subtract means from x and y.
x_zero = x - x_avg
y_zero = y - y_avg

# Dot product of mean-adjusted x and y divided by dot product of mean adjusted x with itself.
m = np.sum(x_zero * y_zero) / np.sum(x_zero * x_zero)
# Subtract m times average x from average y.
c = y_avg - m * x_avg

# Let's have a look - same values as above.
m, c
```

$$ petalwidth = t (sepallength) + u (sepalwidth) + v (petallength) + c $$

```python
# It is so hard to plot it and visualize it, it is not possible, so we can use more complex methods.
```

```python
# Let's pretend we want to do linear regression on these variables to predict petal width.
x = df[['sepallength', 'sepalwidth', 'petallength']]
```

```python
# here is a petal width
y = df['petalwidth']
```

```python
# Asked model to fit the data
m.fit(x, y)
```

```python
# there is intercept.
m.intercept_
```

```python
# The coefficients, in order.
m.coef_
```

```python
# Is it good fit?
m.score(x, y)
```

```python

# Calculating the score by hand.
t, u, v = m.coef_
c = m.intercept_

y_avg = y.mean()

u = ((y - (t * x['sepallength'] + u * x['sepalwidth'] + v * x['petallength'] + c))**2).sum()
v = ((y - y.mean())**2).sum()

1 - (u/v)
```

## Using stats models

```python
# Using statsmodels.
import statsmodels.api as sm

# Tell statmodels to include an intercept.
xwithc = sm.add_constant(x)

# Create a model.
msm = sm.OLS(y, xwithc)
# Fit the data.
rsm = msm.fit()
# Print a summary.
print(rsm.summary())
```
